```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#carregando os dados em um data frame

df = pd.read_csv('/content/vendas_sorvete.csv')
```

```python
#separando as variaveis

temperatura = df[['Temperatura (Â°C)']]
vendas = df['Vendas de Sorvete']
```

```python
#criando o modelo

modelo = LinearRegression()
modelo.fit(temperatura, vendas)
```

```python
#verificando os coeficientes do modelo

print("Coeficiente angular: {}".format(modelo.coef_[0]))
print("Coeficiente linear: {}".format(modelo.intercept_))
```

```python
#visualizando a regressÃ£o

plt.scatter(temperatura, vendas, color="blue", label="Dados reais")
plt.plot(temperatura, modelo.predict(temperatura), color="red", label="RegressÃ£o linear")
plt.xlabel("Temperatura (Â°C)")
plt.ylabel("Vendas de Sorvete")
plt.legend()
plt.title("RelaÃ§Ã£o entre Temperatura e Vendas de Sorvete")
plt.show()
```

```python
#testando uma previsÃ£o

modelo.predict([[30.5]])
```
